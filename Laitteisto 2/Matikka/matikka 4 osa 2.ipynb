```python
import numpy as np
import matplotlib.pyplot as plt

# Define the quadratic function
def trajectory_function(x):
    return -0.0063 * x**2 + 0.55 * x

# Find the vertex (maximum height)
a = -0.0063
b = 0.55
vertex_x = -b / (2 * a)
vertex_y = trajectory_function(vertex_x)

# Find the roots (distance the ball travels)
roots = np.roots([a, b, 0])
distance = max(roots)

# Generate x values for plotting
x_values = np.linspace(0, distance, 400)
y_values = trajectory_function(x_values)

# Plot the function
plt.plot(x_values, y_values, label='y = -0.0063x^2 + 0.55x')
plt.scatter(vertex_x, vertex_y, color='red', zorder=5)  # Mark the peak
plt.text(vertex_x, vertex_y, f'Peak ({vertex_x:.2f}, {vertex_y:.2f})', fontsize=12, verticalalignment='bottom')
plt.scatter(distance, 0, color='blue', zorder=5)  # Mark the distance
plt.text(distance, 0, f'Distance ({distance:.2f}, 0)', fontsize=12, verticalalignment='top')

# Add labels and title
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Ball Trajectory')
plt.legend()
plt.grid(True)
plt.show()

# Print the results
print(f"The ball reaches a maximum height of {vertex_y:.2f} meters.")
print(f"The ball travels a distance of {distance:.2f} meters.")
```

```python
# Define the given values
v1 = 80  # km/h
d1 = 30  # m

# Calculate the proportionality constant k
k = d1 / (v1 ** 2)

# Define the speeds for which we want to calculate the braking distances
speeds = [60, 100, 120]  # km/h

# Calculate the braking distances
distances = {v: k * (v ** 2) for v in speeds}

# Print the results
for v, d in distances.items():
    print(f"The braking distance from {v} km/h is {d:.2f} meters.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Data
F = np.array([1, 2, 3, 5])  # Voima (N)
delta_s = np.array([9, 19, 27, 46])  # Venym (mm)

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(F, delta_s)

# Generate fitted line
F_fit = np.linspace(min(F), max(F), 100)
delta_s_fit = slope * F_fit + intercept

# Plot data and fitted line
plt.scatter(F, delta_s, color='blue', label='Data points')
plt.plot(F_fit, delta_s_fit, color='red', label=f'Fitted line: y = {slope:.2f}x + {intercept:.2f}')
plt.xlabel('Force (N)')
plt.ylabel('Extension (mm)')
plt.title('Extension vs Force')
plt.legend()
plt.grid(True)
plt.show()

# Print the results of the linear regression
print(f"Slope: {slope:.2f} mm/N")
print(f"Intercept: {intercept:.2f} mm")
print(f"R-squared: {r_value**2:.2f}")
```
